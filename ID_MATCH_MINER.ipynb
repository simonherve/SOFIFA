```python
import selenium
PATH = "C:\Program Files (x86)\chromedriver.exe"
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup
import re

import pandas as pd
```

```python
tab = []

for i in range(3900, 4400):
    print(i)
    driver  = webdriver.Chrome(PATH)
    driver.get(f"https://fr.whoscored.com/Matches/146{i}")
    delay = 5
    try:
        myElem = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.ID, "content-header")))
        html = driver.page_source
    except TimeoutException:
        print("Loading took too much time")
    driver.close()
    select = BeautifulSoup(html, "html.parser").find("div", {"id": "breadcrumb-nav"}).find("a").get_text().replace("\n","").replace(" ","")
    if select == "Ligue1-2020/2021":
        tab.append(f"146{i}")
    else:
        print(select)
```

```python
len(tab)
```

```python
tab = pd.DataFrame(tab)
```

```python
tab.to_csv("id_match21.csv")
```
